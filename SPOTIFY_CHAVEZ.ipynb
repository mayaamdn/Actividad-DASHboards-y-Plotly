```python
spotify_palette = ["#1DB954", "#1ED760", "#191414", "#FFFFFF", "#D7DAE5"]
```

```python
import pandas as pd
import altair as alt
spotify = pd.read_csv('spotify.csv')
spotify.head()
```

```python
spotify['Date'] = pd.to_datetime(spotify['Date'])
spotify_long = spotify.melt(id_vars=['Date'], var_name='Song', value_name='Streams')
spotify_long.dropna(subset=['Streams'], inplace=True)
spotify_long.head()
```

## Gr√°fico 1: Streams Diarios de 'Shape of You'


```python
shape_of_you_streams = spotify_long[spotify_long['Song'] == 'Shape of You']

chart = alt.Chart(shape_of_you_streams).mark_line(point=True).encode(
    x=alt.X('Date:T', axis=alt.Axis(title='Fecha')),
    y=alt.Y('Streams:Q', axis=alt.Axis(title='Streams Diarios')),
    color=alt.value(spotify_palette[0]),
    tooltip=[
        alt.Tooltip('Date:T', title='Fecha', format='%Y-%m-%d'),
        alt.Tooltip('Streams:Q', title='Streams', format='.0f')
    ]
).properties(
    title='Streams Diarios de \'Shape of You\''
).interactive()

chart
```

Esta gr√°fica muestra las reproducciones diarias de la canci√≥n 'Shape of You', como se puede observar la canci√≥n fue muy popular a principios del a√±o, sin embargo ha ido disminuyendo.

## Gr√°fico 2: Streams Diarios de 'Despacito'

```python
despacito_streams = spotify_long[spotify_long['Song'] == 'Despacito']

chart = alt.Chart(despacito_streams).mark_line(point=True).encode(
    x=alt.X('Date:T', axis=alt.Axis(title='Fecha')),
    y=alt.Y('Streams:Q', axis=alt.Axis(title='Streams Diarios')),
    color=alt.value(spotify_palette[1]),
    tooltip=[
        alt.Tooltip('Date:T', title='Fecha', format='%Y-%m-%d'),
        alt.Tooltip('Streams:Q', title='Streams', format='.0f')
    ]
).properties(
    title='Streams Diarios de \'Despacito\''
).interactive()

chart
```

Esta gr√°fica muestra las reproducciones diarias de la canci√≥n 'Despacito', como se puede observar la canci√≥n no empezo siendo muy popular a principios del a√±o, sin embargo tuvo un pico en Abril que fue probablemente cuando esta se hizo viral.

## Gr√°fico 3: Comparaci√≥n de las Top 3 Canciones




```python
top_3_songs = ['Shape of You', 'Despacito', 'Something Just Like This']
filtered_spotify = spotify_long[spotify_long['Song'].isin(top_3_songs)]

# Define a custom color scale using the first three colors from spotify_palette
color_scale = alt.Scale(domain=top_3_songs, range=spotify_palette[:3])

chart = alt.Chart(filtered_spotify).mark_line(point=True).encode(
    x=alt.X('Date:T', axis=alt.Axis(title='Fecha')),
    y=alt.Y('Streams:Q', axis=alt.Axis(title='Streams Diarios')),
    color=alt.Color('Song:N', scale=color_scale, legend=alt.Legend(title='Canci√≥n')),
    tooltip=[
        alt.Tooltip('Date:T', title='Fecha', format='%Y-%m-%d'),
        alt.Tooltip('Song:N', title='Canci√≥n'),
        alt.Tooltip('Streams:Q', title='Streams', format='.0f')
    ]
).properties(
    title='Streams Diarios de las Top 3 Canciones'
).interactive()

chart
```

En este gr√°fico se pueden apreciar las 3 canciones mas populares y sus streams, podemos ver que Despacito fue la mas viral en alg√∫n punto pero Shape of You y Something Just Like This duraron mas tiempo siendo populares.
